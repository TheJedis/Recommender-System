```python
import csv
other_input_file = csv.DictReader(open("data/winemag-data-130k-v2.csv", encoding="utf8"))
other_data = []
with open('data/winemag-data-130k-v2.csv', encoding="utf8") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if(row[0]==''):
            continue
        other_data.append(row)
```

```python
fdata = []
for row in other_data:
    if row[10] and row[13]:
        fdata.append( [row[10], row[12], row[4] ] )
print( len(fdata)*100/len(other_data) )
```

```python
import pandas as pd
gdf = pd.DataFrame( fdata ) # Global Data Frame
```

```python
import numpy as np
msk = np.random.rand(len(gdf)) < 0.8
train = gdf[msk]
test = gdf[~msk]
test = test.iloc[:,0:2]
```

```python
from surprise import Reader, Dataset
train_reader = Reader(line_format='user item rating')#, sep='\t'
test_reader = Reader(line_format='user item')#, sep='\t'
traindata = Dataset.load_from_df(train, reader=train_reader)
testdata = Dataset.load_from_df(test, reader=test_reader)
```

```python
from surprise import SVD
algo = SVD()
```

```python
trainset = traindata.build_full_trainset()
algo.fit(trainset)
```

```python
testset = testdata.build_full_trainset()
testset = testset.build_testset()
```

```python
algo.test( testset )
```
